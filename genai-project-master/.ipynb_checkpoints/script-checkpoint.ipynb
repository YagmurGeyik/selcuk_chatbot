```python
from pymilvus import Collection, FieldSchema, CollectionSchema, DataType,connections,utility
```

```python
import json
```

```python
import openai
```

```python
openai.api_key = ""
```

```python
connections.connect(
    alias="default",
    host="localhost",
    port="19530"
)
```

```python
json_file = 'updated_format.json'
with open(json_file, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
with open('updated_' + json_file, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```

```python
item_id = FieldSchema(
    name="item_id",
    dtype=DataType.INT64,
    auto_id=True,
    is_primary=True,
)
```

```python
headers = FieldSchema(
    name="header",
    dtype=DataType.VARCHAR,
    max_length=255
)
```

```python
contexts = FieldSchema(
    name="context",
    dtype=DataType.VARCHAR,
    max_length=64000
)
```

```python
vector_contexts = FieldSchema(
    name="vector_context",
    dtype=DataType.FLOAT_VECTOR,
    dim=1536
)
```

```python
header = [item['header'] for item in data]
```

```python
context = [item['context'] for item in data]
```

```python
vector_context = [item['vector_context'] for item in data]
```

```python
collection_schema=CollectionSchema(
    fields=[item_id,headers,contexts,vector_contexts],
    description="rules_qa",
)
```

```python
utility.drop_collection("rules_qa")
```

```python
collection = Collection(
    name="rules_qa",
    schema=collection_schema,
    using="default"
)
utility.list_collections()
```

```python
mr = collection.insert([
    header,
    context,
    vector_context
])
```
